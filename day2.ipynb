```python
# 사용자가 자연수를 입력하면 입력받은 자연수를 누적합 후 출력
# 단, 음수를 입력하면 입력 종료
total = 0

# break 키워드 
# 가장 가까운 반복문을 탈출한다

# 조건식 자리에 True -> 무한 반복문 : 반드시 탈출 조건을 명시할 것.


while True:
    num = int(input("자연수를 입력: "))
    
    if num<0:
        break

    total += num
    print(f"누적 합 : {total}")
```
