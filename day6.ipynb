# 머신러닝 기초(Machine Learning Basics)
- 머신러닝이 주목받는 이유 : 데이터의 폭증, 컴퓨팅 파워의 발전
- 핵심 용어
- Feature(특성) : Model의 입력 값(X), 공부 시간
- Label(레이블/정답) : Model이 예측해야 될 값(y), 취업률
- Training(학습) : Feature 와 Label 데이터를 이용해 Model을 만드는 과정
- Inferrence(추론) : 학습된 Model을 사용해 새로운 데이터의 Label을 예측하는 과정

- 머신 러닝의 종류
- 지도 학습(Supervised Learning) : Label이 있는 데이터로 학습
- 예) 스팸 메일 분류, 주택 가격
- 비지도 학습(Unsupervised Learning) : Label이 없는 데이터로 학습 -> 패턴 학습
- 예) 마케팅 고객 그룹 분류, 이상 탐지 모델
- 강화 학습(Reinforcement Learning) : 보상(Reward)을 최대화하는 방향으로 학습
- 예) Alpha-Go, 자율주행차

# 데이터 분리
- Train 데이터와 Test 데이터로 나누자
- why? 모델이 학습 과정 중 외운 정답을 통해 답을 예측하는 것을 방지.
- [2.5,3.1,4.,0.2] -> 내 모델 -> 예측
- 어? 이거 봤던건데 ? -> 이거 정답은 0 이야
- 우리는 sklearn의 train_test_split 함수 사용 아니면 랜덤 라이브러리 사용

```python
from sklearn.model_selection import train_test_split  #train, test 분리

X = iris_dataset.data
y = iris_dataset.target

# train, test의 비율을 75:25 비율로 분할 >어떤비율로 나누는지도 정확도에 영향이 감.
# scatter 산점도에서(day5),,, 랜덤 데이터인데 시드를 설정하면 같은 결과가 나오게 됨
# (X_train, X_test, y_train ,y_test) 의 tuple 형태로받음
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=37)

print("X_train의 크기 :", X_train.shape)
print("X_test 크기 :", X_test.shape)
print("y_train 크기 :", y_train.shape)
print("y_test 크기 :", y_test.shape)

print("학습용 데이터의 일부 : \n", X_train[:3])
```

```python
# 1)input 과 Label을 통해 머신 러닝 없이도 해결할 수 있는 문제인가 ?
# input Label
#   1     3
#   2     5
#   3     7
#  y= 2X + 1
# 2) 산점도 시각화를 사용해서 실제로 분류할 수 있는가를 확인

import pandas as pd

#X_train 활용 df 만들기
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
# iris_dataframe.describe()
import mglearn
# 혹시나 X_test,y_test는 사용 X (평가용으로만 사용 예정)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker="o",
                            hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
# 첫 번째 머신러닝 모델 선정(KNN)
# K-최근접-이웃(K-Nearest Neighbors)
# 가장 직관적인 머신 러닝 모델
# 나는 누구인가 ? -> 내 이웃을 보고 결정하자
# 모델 - >(K=3, 정확도 70%, K=5, 정확도75%, K=7,정확도 73%)
# K 값을 조정하는 과정 -> prarameter tuning 한다
# K ": Hyper Parameter" 
from sklearn.neighbors import KNeighborsClassifier
# 모델 객체 생성(K=n) (n값은 나중에 다른 값으로 조정해보기--7일때 정확도 떨어짐)
knn = KNeighborsClassifier(n_neighbors = 3)
#모델 학습 : fit
knn.fit(X_train,y_train)
```

```python
# 학습된 결과를 바탕으로 평가
# y_test
# X_test -> knn -> y_pred
# y_pred와 y_test 값을 비교해서
# 정확도(accuracy)를 확인
y_pred = knn.predict(X_test)
print("우리가 만든 모델로 예측한 결과 : \n",y_pred)
print("실제 Label : \n",y_test)

# 정확도 측정 라이브러리
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)
print(f"예측 모델 정확도 : {accuracy*100:.2f}%")
```

```python
# K값에 따라 정확도 변화를 확인해보자 (k=1,3,5,7,...)
k_list = range(1,30,2)
accuracies = []
for k in k_list:
    knn_test = KNeighborsClassifier(n_neighbors=k)
    knn_test.fit(X_train,y_train)
    acc = accuracy_score(y_test,knn_test.predict(X_test))
    accuracies.append(round(acc,3))
    
print(accuracies)
```

```python
# 모델의 성능
#과소적합(Underfitting) : 모델이 너무 단순해서 데이터의 패던을 제대로 파악하지 못함
# -> 학습 성능 낮다, 테스트 성능 낮다 (:>>높다가 아닌가 ?)
#과적합(Overfitting) : 모델이 Train데이터에 너무 의존하여서 새로운 데이터에 대해 제대로 예측하지 못하는 경우
# -> 학습 성능 높다, 테스트 성능 낮다
# 일반화(Generalization) : 모델이 학습뿐만 아니라 새로운 데이터에 대한 정답도 잘 도출
# -> 이상적인 목표

# 과소적합/과적합 어떻게 확인할 것인가 ?
# 학습 데이터셋, 테스트 데이터셋 간의 성능 차이 확인
# 해결법 
# 과소적합일 경우 : 더 복잡한 모델을 사용, input의 특성의 갯수를 늘림(Feature 추가) : 중고차 구매시 차량 색상추가
# 과적합일 경우 : 더 많은 데이터를 확보한다, 특정 Feature이 과도한 영향을 미치는 경우 Feature를 제한(Regulariztion),다른 케이스에 대해 검증을 추가진행 (교차검증:Cross-Validation)
```

```python
# 지도학습을 선택하는 경로
# input,Label 
# 정답(Label)이 있으니까 지도학습을 해야겠다.
# 그럼 분류를 할 것인가 회귀를 할 것인가 ?

# 분류(Classification)
# 붓꽃의 3품종에 관한 분류 -> 다중 분류(class가 여러개다)
# 이진분류 : 스팸 메일인가 아닌가? / 강아지인가 고양이인가?
# 붓꽃의 정답이(Label)이 이산적인 값

# 회귀(Regression)
# 어떤 연속적인 값을 예측
# Label의 값이 연속적인 값.
```

```python
# 2번째 분류 알고리즘
# 의사 결정 트리(Decision Tree) - ex)아키네이터
# 원리 : Feature에 대한 '질문'을 통해서 데이터를 분류
# 핵심 개념

# -불순도(Impurity) : 한 노드에 여러 클래스의 데이터가 얼마나 섞여 있는지에 관한 척도
# --지니 불순도(Gini Impurity) : 클래스가 섞여 있지 않을 수록  0
# ex)상자에 빨간공5 파란공5 , 공을 2개 꺼낼 때 동일한 색이 나올수록 순수하다고 표현, 다른 색이 나오면 불순
# 위의 경우 5/9가 Gini불순도(약 0.55)
# --엔트로피(Entropy) : 지니 불순도와 비슷한 개념 (클래스들이 얼마나 안정되어 있는지에 관한 척도)
# 지니불순도가 엔트로피보다 계산이 빠름

# -정보획득(Information Gain) : 특정 질문을 통해 데이터를 나누었을 때 불순도가 얼마나 감소하는지를 측정하는 과정

# 의사 결정 트리 -> 정보 획득이 최대가 되는 방향으로 학습된다. (불순도를 낮추는 방향으로 진행)
# 장점 : 어떤 규칙에 의거해서 분류하는지 눈으로 보기 명확함,
#        사람의 언어로 설명하기 용이함, 시각화하기 쉽다., 전처리가 쉽다.(데이터 스케일에 영향을 덜 받는다.)
# 단점 : ★과적합 될 확률이 높다.(학습한 데이터에 대해서 잘 맞추지만, 실제 데이터에 대한 예측은 낮을 수 있다.)
#       랜덤하게 나눈 train data가 변하면 의사결정 모델이 급격히 변화한다(데이터 자체에 대한 영향이 있음)

# **과적합을 방지하기 위해** depth를 제한, 노드에 들어가는 최소 데이터 수를 제한하여 Overfitting를 방지
```

```python
# 의사 결정 트리 활용한 와인 데이터 분류 (교재p106)
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split # 이 셀을 실행했을때만 (X,y가 위와 중복 변수라서)
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

wine = load_wine() # wine 데이터 로드

X = wine.data
y = wine.target

# 데이터 분할 train:test =0.75:0.25
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=6)
# max depth : 과적합(Overfitting)을 방지하기 위한 parameter
d_tree = DecisionTreeClassifier(max_depth=3, random_state=6) 
d_tree.fit(X_train, y_train)

# 학습된 결과를 통해 평가
y_pred = d_tree.predict(X_test)
accuracy_score(y_test,y_pred) # 둘을 비교해 정확도 
print(f"정확도 : {accuracy:.4f}")
```

```python
# 의사 결정 모델 시각화
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(20,15))
plot_tree(d_tree, feature_names = wine.feature_names,class_names= wine.target_names,filled=True)
plt.show()
```

```python
# 정보 획득
# 어떤 요인(feature)이 가장 영향을 미쳤는가?
feature_importances = d_tree.feature_importances_
df_wine = pd.DataFrame({"feature":wine.feature_names,"importances":feature_importances})
df_wine
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

cancer = load_breast_cancer()
print(cancer.data.shape)

X = cancer.data
y = cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)


tree = DecisionTreeClassifier(max_depth=3 ,random_state=0) # max_depth=n 원래 학습 데이터 내에서는 정확도가 낮아져도 
                                              # 외부에서 새로운 데이터에 대한 정확도가 소폭 상승함

tree.fit(X_train,y_train)

# 의사 결정 나무 안에 정확도 측정 함수
# 모델을 다 만든 후 원래 학습 했었던 X, y를 넣어보자
print(f"학습 세트 정확도: {tree.score(X_train,y_train):.3f}")
print(f"테스트 세트 정확도: {tree.score(X_test,y_test):.3f}")


# 정확도 측정
y_pred = tree.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f"정확도 : {accuracy:.3f}")
```

```python
# 유방암 모델 의사 결정 트리 시각화
from sklearn.tree import plot_tree
# import matplotlib.pyplot as plt //재시작 할 떄 plt 없다면 실행
plt.figure(figsize=(20,15))
plot_tree(tree,filled=True, feature_names=cancer.feature_names,class_names=cancer.target_names)
plt.show()

#depth가 많다고 무조건 좋은 것이 아님. 과적합가능성
#depth가 많아질수록 실행속도 증가
```

```python
# SVM (Support Vector Machine) :교재 p144
# 각 클래스의 데이터를 가장 잘 구분하는 경계선(Decision Boundary)을 찾는 알고리즘
# 핵심 개념
# -마진(Margin) : 결정 경계와 가장 가까운 각 클래스의 데이터 샘플(Support Vector)사이의 거리
# -서포트 벡터(Support Vector) : 결정 경계를 정의하는 데 직접적인 영향을 주는 데이터
# 특징 : 분류 문제에서 굉장히 강력한 성능을 보임. 
```

```python
# 붓꽃(iris)데이터를 활용한 SVM 모델 기반 학습
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # 전처리 위한 클래스

from sklearn.svm import SVC

iris_dataset = load_iris()
X = iris_dataset.data
y = iris_dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=37)

# 스케일링 (feature 간 범위 폭을 일정하게 조절하는 것)
# 학습 하기 전에 조절하는 것 > (전처리)

scaler = StandardScaler()
#X_train의 feature가 3개다 -1 ~ 0 ~ 1사이 표준편차 : 1
X_train_scaled = scaler.fit_transform(X_train) # X_train을 살피면서 평균과 표준편차 등을 학습
X_test_scaled = scaler.transform(X_test) #학습된 평균, 표준편차 등을 토대로 X_test를 스케일링


# 1번이 1-10 2번이 1-1000일때 스케일링이 2번을 100으로 나눈다고해서 되는 것이 아님. 평균이나 표준편차를 고려해야함
# 이를 적절히 조절해서 스케일링 하는것


# kernel = "rbf" : 비선형 데이터에 사용되는 커널
# c : 규제 파라미터
svm = SVC(kernel="rbf",C=1.0,random_state=37)
svm.fit(X_train_scaled,y_train)
```

```python
# 정확도 평가
from sklearn.metrics import accuracy_score
y_pred = svm.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"정확도 : {accuracy:.4f}")
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import graphviz

# 1. iris 데이터 로드
iris_dataset = load_iris()
X = iris_dataset.data
y = iris_dataset.target

# 2. train, test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=37)

# 3. 의사결정트리 모델 학습(fit)
iris_tree = DecisionTreeClassifier(max_depth=3 ,random_state=37) 
iris_tree.fit(X_train,y_train)

# 4. 예측 값과 테스트 label 비교 -> 정확도 측정
y_pred = iris_tree.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f"정확도 : {accuracy:.4f}")

# 5. export_graphviz 함수 사용해서 "iris.dot"export
export_graphviz(iris_tree, feature_names=iris_dataset.feature_names, class_names=iris_dataset.target_names,
                out_file="iris.dot",impurity=False, filled= True)

with open("iris.dot", "r", encoding="UTF-8") as file:
    dot_graph = file.read()

# 6. graphviz 시각화
display(graphviz.Source(dot_graph))
```

```python
# 지도 학습
# 회귀(Regression)
# 선형 회귀(Linear Regression) : 데이터에 가장 잘 맞는 '직선' 을 찾는 알고리즘 (독립 변수 x 와 종속변수 y의 선형관계)
# 기본 모델 : y = Wx + b (W: 가중치, b: 절편)
# 실제 y_test 값과 input을 통해 얻은 y(wx + b) 값의 평균 제곱 오차(MSD:Mean Squared Error)를 사용 
# 해당 함수들을 -> 비용 함수 (Cost Fucntion) : 실제 값과 예측 값의 오차를 측정하는 함수
# MSE가 최소화하는 방향으로 진행
# 만약 편미분 했을 경우 0이 되는 값을 찾지 못했을 때, 어떤 방식으로 W와 b를 수정해 나갈것인가?
# 가장 많이 사용하는 방법 : 경사 하강법 (Gradient Descent)
```

```python
# 실습형 선형 회귀 데이터 생성
from sklearn.datasets import make_regression # 실습용 선형회귀 만드는 함수
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

X,y = make_regression(n_samples=1000,n_features=8,noise=8,random_state=42)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 학습 모델
linear_reg = LinearRegression()
linear_reg.fit(X_train,y_train)

# 예측 결과
y_pred = linear_reg.predict(X_test)

# Cost Function : MSE
mse = mean_squared_error(y_test,y_pred)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score


# mae
mae = mean_absolute_error(y_test,y_pred)
# mse
mse = mean_squared_error(y_test,y_pred)
# rmse
rmse = root_mean_squared_error(y_test,y_pred)
# r2
r2 = r2_score(y_test,y_pred)

print(f"mae : {mae}")
print(f"mse : {mse}")
print(f"rmse : {rmse}")
print(f"r2 : {r2}") # 1에 가까울수록 좋음


# 제곱이기 때문에 mse 값이 너무 커지게 될 경우가 있음 -> root를 씌운다
# rmse: mse에 root를 씌운다

import numpy as np
rmse = np.sqrt(mse)

#가중치(W)와 y절편(b)를 구해보자.
print("가중치 W :",linear_reg.coef_)
print("y절편 b :", linear_reg.intercept_)
```

# 회귀의 평가지표
# 연속적인 데이터이므로 '수치의 차이'를 통해 평가한다
# W,b -> 구한 예측값 (y_pred)과 원래값(y_test)의 차이가 작을수록 좋은 모델임.
# 오차의 차이만 가지고 진행 할 경우 +,- 가 상쇄될 가능성이 있음 
# 따라서 절댓값, 제곱을 통해 양의 범위에서 처리할 것

* 1) MAE(Mean Absolute Error) : 절댓값 차이 평균
 * $MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|$ (오차 절대값의 평균)

* 2) MSE(Mean Squared Error) : 제곱 차이 평균
* -> MAE 대비 멀리 떨어질 수록 패널티를 가중
 * $MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$ (오차 제곱의 평균, 큰 오차에 패널티 부여)

* 3) RMSE (Root Mean Squared Error) : MSE에 루트를 씌워 실제 값과 단위를 맞춤 
 * $RMSE = \sqrt{MSE}$ (MSE에 루트를 씌워 실제 값과 단위를 맞춤)

* 4) R^2 (결정 계수, Coefficient of Determination) : 모델이 데이터의 분산을 얼마나 잘 설명하는지에 관해서 나타내는 지표 (1에 가까울수록 좋은 모델임.)
 * $R^2 = 1 - \frac{\sum(y_i-\hat{y}_i)^2}{\sum(y_i-\bar{y}_i)^2}$

```python
from sklearn.datasets import make_regression # 실습용 선형회귀 만드는 함수
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import numpy as np

from sklearn.datasets import load_diabetes
X, y = load_diabetes(return_X_y=True)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 학습 모델
linear_regress = LinearRegression()
linear_regress.fit(X_train, y_train)

# 예측 결과
y_pred = linear_regress.predict(X_test)

# 평가 지표
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score


# mae
mae = mean_absolute_error(y_test,y_pred)
# mse
mse = mean_squared_error(y_test,y_pred)
# rmse
rmse = root_mean_squared_error(y_test,y_pred)
# r2
r2 = r2_score(y_test,y_pred)

print(f"mae : {mae}")
print(f"mse : {mse}")
print(f"rmse : {rmse}")
print(f"r2 : {r2}") # 1에 가까울수록 좋음



# 가중치(W)와 y절편(b)를 구해보자
print("가중치 W :", linear_regress.coef_)
print("y절편 b :", linear_regress.intercept_)
```
