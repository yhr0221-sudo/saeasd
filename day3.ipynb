```python
# 정렬 알고리즘
# list의 가지고 있는 요소들을 순서대로 열거하는 알고리즘
array=[45,7,12,82,25]
print("정렬 전 :",array)
#오름차순으로 정렬
# Bubble Sort (버블정렬): 두 인접한 요소의 크기를 비교 , 치환함으로써 정렬
#step 1
# if array[0]>array[1]:
#     temp = array[1]
#     array[0] = array[1]
#     array[1] = temp
# if array[1] >array[2]:
#     temp = array[1]
#     array[1] = array[2]
#     array[2] = temp

# if array[2] >array[3]:
#     temp = array[2]
#     array[2] = array[3]
#     array[3] = temp

for i in range(4):
    if array[i]>array[i+1]:
        temp = array[i] 
        array[i] = array[i+1] 
        array[i+1] = temp

print("정렬 후 :", array) 

#step 2
#array[0]이 array[1]보다 크다면 치환
#array[1]이 array[2]보다 크다면 치환
#array[2]이 array[3]보다 크다면 치환

for i in range(3):
    if array[i]> array[i+1]:
        temp = array[i]
        array[i] = array[i+1]
        array[i+1] = temp
print("정렬 후 :", array) 

# 위에서 규칙을 찾은 뒤 정리 한 식
for j in range(4):
    for i in range(4 - j):  # 4>3>2>1 순서로 쭉 진행 -> array[4] 자리부터 값이 채워짐 > 오름차순
        if array[i] > array[i + 1]:
            temp = array[i]
            array[i] = array[i + 1]
            array[i + 1] = temp

print("정렬 후 :", array) # 7, 12, 25, 45, 82
```

```python
# 선택 정렬 ; step당 치환을 1번만 함.
# Selection Sort > 가장 큰 값 혹은 가장 작은 값의 위치를 찾아 교환함으로써 정렬

a=[98,7,70,13,24]
print("선택 정렬 전 :",a)

#Logic 오름차순 정렬
#step 1 
max_index=0 # 가장 큰 값의 위치 정보
# 규칙 찾기
# if a[0] > a[max_index]:
#     max_index = 0
# if a[1] > a[max_index]:
#     max_index = 1
# if a[2] > a[max_index]:
#     max_index = 2

for i in range(5):
    if a[i] > a[max_index]:
        max_index = i

# 탐색 후 치환
temp = a[max_index]
a[max_index] = a[4]
a[4] = temp
print("선택 정렬 후 :", a)
#step2
max_index = 0 
for i in range(4):
    if a[i] > a[max_index]:
        max_index = i
temp = a[max_index]
a[max_index] = a[3]
a[3] = temp

print("선택 정렬 후 :", a)

#step 3 
max_index = 0 
for j in range(4):
    for i in range(5 - j):
        if a[i] > a[max_index]:
            max_index = i
    temp = a[max_index]
    a[max_index] = a[4-j]
    a[4-j] = temp
    
print("선택 정렬 후 :", a)
```

```python
# 사용자 정의 함수
#1) 매개변수가 존재, 리턴값(반환값) 존재
def add_two_number(num1,num2): #num1,num2를 매개변수(parameter)라 함.
    return num1+num2 #반환값
print(add_two_number(10,20)) # 10,20을 전달인자 (argument)라 함. 

#2) 매개변수가 존재, 리턴값이 존재 X
def print_add(num1,num2): # >>>함수의 정의부 def 부분
    print(num1+num2)    
print_add(20,30) # >>>함수의 호출부 (call) print 부분
print_add(30,40)

#1번과 2번의 차이
print(add_two_number(3,5) + add_two_number (10,30))
# print_add(3,5) + print_add(10,30)  >> print() + print() 와 같은 Error가 발생
# 추가적인 연산이 불가능

#3) 매개변수 존재 X , 리턴값 존재
def get_name(): #getter 나중에 추가설명
    return"류이현"
print(get_name())

#4) 매개변수 존재 X, 리턴값 존재 X
def today_seoul_weather():
    print("우중충")
today_seoul_weather() # 호출만 하고 끝날 때 사용
```

```python
# abs
def absolute(num):
    #절댓값 반환
    if num < 0:
       return num*(-1)
    else:
        return num
print(absolute(-990))
print(absolute(10))
```

```python
list5=[10,5,7,20,3]
#list 의 가장 큰 값을 반환하는 maximize정의
def maximize(list00):
    max_value = list00[0]
    for num in list00:
        if num > max_value:
            max_value = num
    return max_value
print(maximize(list5))
```

```python
#close10(num1,num2)
#둘 중에 10 에 더 가까운 수를 return하는
def close10(a,b):
    if absolute(10-a) > absolute(10-b):
        return b
    else:
        return a
num1 = int(input("첫 번째 정수 입력 :"))
num2 = int(input("두 번째 정수 입력 :"))
print(close10(num1,num2))
```
